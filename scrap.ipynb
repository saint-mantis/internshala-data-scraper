```python
from selenium.common.exceptions import NoSuchElementException
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
```

```python
driver = webdriver.Chrome()
driver.get("https://internshala.com/internships/page-1/")
```

```python
rows_needed = 100
n = 0
indexnum = 0
df = pd.DataFrame(columns=['Job','Company','Work Location','Start Date','Duration','Stipend','Time','Internship Type'])

while True:
    target = driver.find_elements(By.CSS_SELECTOR, 'div[class="container-fluid individual_internship  "]')
    for i in target:
        try:
            job = i.find_element(By.CSS_SELECTOR,'div[class="container-fluid individual_internship  "] div[class="internship_meta"] div[class="individual_internship_header"] div[class="heading_4_5 profile"] a[class="view_detail_button"]')
            df.loc[indexnum,'Job'] = job.text
        except:
            df.loc[indexnum,'Job'] = "None"

        try:
            company = i.find_element(By.CSS_SELECTOR,"div[class='company_and_premium']")
            df.loc[indexnum,'Company'] = company.text
        except:
            df.loc[indexnum,'Company'] = "None"

        try:
            work_location = i.find_element(By.CSS_SELECTOR,'div[class="individual_internship_details  individual_internship_internship"] div[id="location_names"] span')
            df.loc[indexnum,'Work Location'] = work_location.text
        except:
            df.loc[indexnum,'Work Location'] = "None"

        try:
            start_date = i.find_element(By.CSS_SELECTOR,'div[class="item_body"] span[class="start_immediately_desktop"]')
            df.loc[indexnum,'Start Date'] = start_date.text
        except:
            df.loc[indexnum,'Start Date'] = "None"

        try:
            duration = i.find_element(By.CSS_SELECTOR,'div[class="internship_meta"] div[class="individual_internship_details  individual_internship_internship"] div[class="internship_other_details_container"] div[class="other_detail_item_row"] div:nth-child(2) div[class="item_body"]')
            df.loc[indexnum,'Duration'] = duration.text
        except:
            df.loc[indexnum,'Duration'] = "None"

        try:
            stipend = i.find_element(By.CSS_SELECTOR,'div[class="other_detail_item  stipend_container"] div[class="item_body"] span[class="stipend"]')
            df.loc[indexnum,'Stipend'] = stipend.text
        except:
            df.loc[indexnum,'Stipend'] = "None"

        try:    
            timee = i.find_element(By.CSS_SELECTOR,'div[class="posted_by_container"] div[class="status-container"] div[class="status status-small status-inactive"] ')
            df.loc[indexnum,'Time'] = timee.text
        except:
            df.loc[indexnum,'Time'] = "None"

        try:
            internship_type = i.find_element(By.CSS_SELECTOR,'div[class="other_label_container"] div[class="status-container"] div[class="status status-small status-inactive"] ')
            df.loc[indexnum,'Internship Type'] = internship_type.text
        except:
            df.loc[indexnum,'Internship Type'] = "None"

        print(n)
        n+=1
        indexnum+=1
        if n == rows_needed:
            break
        else:
            continue
    if n == rows_needed:
        break
    else:
        next_btn = driver.find_element(By.CSS_SELECTOR,'div[class="internship_list_container"] nav div[id="pagination"] a[class="next_page hideUndoOnClick"]').click()
        time.sleep(5)
        continue
    
```
